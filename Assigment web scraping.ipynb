```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd

driver = webdriver.Chrome()
driver.get(url="https://wuzzuf.net/jobs/egypt")
search_box = driver.find_element(By.NAME, "q")
search_box.clear()
search_box.send_keys("software engineering")
search_box.send_keys(Keys.RETURN)
time.sleep(5)

job_links = []
page_number = 1
max_pages = 2
while page_number <= max_pages:
    try:
        jobs = driver.find_elements(By.XPATH, "//a[contains(@href,'/jobs/p/')]")
        for j in jobs:
            link = j.get_attribute("href")
            if link not in job_links:
                job_links.append(link)
        try:
            next_button = driver.find_element(By.XPATH, "//button[@class='css-wq4g8g ezfki8j0']")
            next_button.click()
            time.sleep(5)
            page_number += 1
        except:
            break
    except Exception as e:
        print(f"Error on page {page_number}: {e}")
        break

print(f"Total job links collected: {len(job_links)}")

job_data = []
for idx, link in enumerate(job_links, 1):
    try:
        driver.get(link)
        job_info = {}
        try:
            job_info['Job Title'] = driver.find_element(By.XPATH, "//h1").text.strip()
        except NoSuchElementException:
            job_info['Job Title'] = None
        try:
            job_info['Location'] = driver.find_element(By.XPATH, "//span[contains(@class,'css-5wys0k')]").text.strip()
        except NoSuchElementException:
            job_info['Location'] = None
        try:
            job_info['Experience'] = driver.find_element(By.XPATH, "//div[contains(text(),'Exp')]").text.strip()
        except NoSuchElementException:
            job_info['Experience'] = None
        try:
            job_info['Date Posted'] = driver.find_element(By.XPATH, "//div[contains(@class,'css-4c4ojb')]").text.strip()
        except NoSuchElementException:
            job_info['Date Posted'] = None
        try:
            job_info['Job Type'] = driver.find_element(By.XPATH, "//span[contains(@class,'css-1ve4b75')]").text.strip()
        except NoSuchElementException:
            job_info['Job Type'] = None
        try:
            skills_elems = driver.find_elements(By.XPATH, "//a[contains(@class,'css-5x9pm1')]")
            skills = [s.text.strip() for s in skills_elems]
            job_info['Skills'] = ", ".join(skills)
        except NoSuchElementException:
            job_info['Skills'] = []

        job_info['Application Link'] = link
        job_data.append(job_info)

        
        print(f"Scraped job {idx}/{len(job_links)}:")
        print(job_info)
        print("-*-" * 25)

    except Exception as e:
        print(f"Error scraping {link}: {e}")

df = pd.DataFrame(job_data)
df.to_csv("wuzzuf_jobs.csv", index=False, encoding="utf-8-sig")
df.to_json("wuzzuf_jobs.json", orient="records", force_ascii=False)

print("Jobs saved to wuzzuf_jobs.csv and wuzzuf_jobs.json")
```
